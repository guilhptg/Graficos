```python
import pandas as pd
import matplotlib.pyplot as plt


paises_df = pd.read_csv('Olympics_Country.csv')
jogos_df = pd.read_csv('Olympic_Games_Medal_Tally.csv')
resultados_altle_df = pd.read_csv('Olympic_Athlete_Event_Results.csv')
games_medal_df = pd.read_csv('Olympic_Games_Medal_Tally.csv')
games_df = pd.read_csv('Olympics_Games.csv')

display(games_medal_df)
display(games_df)
```

# Medalhas

```python
games_medal_2022_df = games_medal_df[games_medal_df['edition'] == '2022 Winter Olympics']
country = games_medal_2022_df[['country_noc']]
country_name = games_medal_2022_df[['country']]
gold = games_medal_2022_df[['gold']]
plt.plot(country['country_noc'], gold['gold'], marker='o')
plt.xlabel('Country')
plt.ylabel('Gold Medal')
plt.grid(True)
plt.show()
```

# Barras

```python

silver = games_medal_2022_df[['silver']]
bronze = games_medal_2022_df[['bronze']]

barras = plt.bar(country['country_noc'], silver['silver'], color='gray')
plt.bar(country['country_noc'], bronze['bronze'], color='red')
plt.bar(country['country_noc'], gold['gold'], color='yellow')
plt.bar_label(barras,labels=country, padding=2, fontsize=9)
plt.xticks(ticks=country['country_noc'])
plt.xlabel('Countrys')

plt.ylabel('Medals')
plt.grid(True)
plt.show()
```

# Pizza / Pie

```python
plt.figure(figsize=(18,5))
plt.plot(country['country_noc'], silver['silver'], color='gray', label='Siver', marker='o', scalex=True, linewidth=3, markersize=4)
plt.plot(country['country_noc'], gold['gold'], color='yellow', label='Gold', marker='o', scalex=True, linewidth=3, markersize=4)
plt.plot(country['country_noc'], bronze['bronze'], color='brown', label='Bronze', marker='o', scalex=True, linewidth=3, markersize=4, animated=True)
plt.ylim(0, 18)
plt.title('Medalhas Olimpiadas - 2022')
plt.ylabel('Medalhas (2022)')
plt.legend()
plt.show()
```
